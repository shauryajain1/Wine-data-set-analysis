```python
df['fixed acidity'].fillna(int(df['fixed acidity'].mean()))
df['volatile acidity'].fillna(int(df['volatile acidity'].mean()))
df['citric acid'].fillna(int(df['citric acid'].mean()))
df['residual sugar'].fillna(int(df['residual sugar'].mean()))
df['chlorides'].fillna(int(df['chlorides'].mean()))
df['free sulfur dioxide'].fillna(int(df['free sulfur dioxide'].mean()))
df['total sulfur dioxide'].fillna(int(df['total sulfur dioxide'].mean()))
df['density'].fillna(int(df['density'].mean()))
df['pH'].fillna(int(df['pH'].mean()))
df['sulphates'].fillna(int(df['sulphates'].mean()))
df['alcohol'].fillna(int(df['alcohol'].mean()))
df['quality'].fillna(int(df['quality'].mean()))
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,annot=True)
```
